```python
#company eke category tika balanwa
data['Company'].value_counts()
```

```python
# 10 ta aduwen thiyna ewa other kiyla column ekakata dagnnawa
def add_company(inpt):
    if inpt == 'Samsung' or inpt == 'Mediacom' or inpt == 'Razer' or inpt == 'Microsoft' or inpt == 'Vero' or inpt == 'Xiaomi' or inpt == 'Chuwi' or inpt == 'Fujitsu' or inpt == 'Google' or inpt == 'LG' or inpt == 'Huawei':
        return 'Other'
    else:
        return inpt

data['Company'] = data['Company'].apply(add_company)
```

```python
# meke godak wisthara thiynwa. ethanin Screen eka IPS hri Touch screen harida kiyla withrak blnwa
data['Touchscreen'] = data['ScreenResolution'].apply(lambda x:1 if 'Touchscreen' in x else 0)
data['Ips'] = data['ScreenResolution'].apply(lambda x:1 if 'IPS' in x else 0)
```

```python
#CPU eke count tika balanwa
data['Cpu'].value_counts()
```

```python
len(data['Cpu'].value_counts())
```

```python
#CPU name godak thiynwa ethakota One hot encoding krnna ba. enisa e warga tika adu krgnnawa wenama column ekak hdgena
#mekedi cpu eke palaweni wachana 3 withrai ganne example widihata Intel Core i5 7200U 2.5GHz kiyna eke Intel Core i5 kiyla gnne
data['cpu_name'] = data['Cpu'].apply(lambda x:" ".join(x.split()[0:3]))
```

```python
data['cpu_name'].value_counts()
```

```python
#mekedi krnne Intel Core i7 ,Intel Core i5  ,Intel Core i3, AMD, Other kiyla catogery wen krgnnawa
def set_processor(name):
    if name == 'Intel Core i7' or name == 'Intel Core i5' or name == 'Intel Core i3':
        return name
    else:
        if name.split()[0] == 'AMD':
            return 'AMD'
        else:
            return 'Other'

data['cpu_name'] = data['cpu_name'].apply(set_processor)
```

```python
data['cpu_name'].value_counts()
```

```python
#GPU eke count tika balanwa
data['Gpu'].value_counts()
```

```python
len(data['Gpu'].value_counts())
```

```python
# GPU eke palaweni nama withrak aran aluth column ekak hdnwa
data['gpu_name'] = data['Gpu'].apply(lambda x:" ".join(x.split()[0:1]))
data['gpu_name'].value_counts()
```

```python
data.shape
```

```python
# me ARM gpu eken ekai thiynne data. eken wdak nathi nisa e data set eka ain krnwa
data = data[data['gpu_name']!='ARM']
data.shape
```

```python
data.head(2)
```

```python
#OS eke count tika balanwa
data['OpSys'].value_counts()
```

```python
# methanadith windows tita wenmath mac tika wenamth linux tika wenmth anthi ewa others wlatath dala aluth column ekak hdanwa
def set_os(inpt):
    if inpt == 'Windows 10' or inpt == 'Windows 7' or inpt == 'Windows 10 S':
        return 'Windows'
    elif inpt == 'macOS' or inpt == 'Mac OS X':
        return 'Mac'
    elif inpt == 'Linux':
        return inpt
    else:
        return 'Other'

data['OpSys'] = data['OpSys'].apply(set_os)
```

```python
data['OpSys'].value_counts()
```

```python
data.head(2)
```

```python
# one nathi ewai apita one widiyata hadgaththa ewge columns tikai dlt krnwa
data = data.drop(columns=['laptop_ID','Inches','Product','ScreenResolution','Cpu','Gpu'])
```

```python
data.head()
```

```python
# dam mulu data set ekama one hot encoding kranwa. one hot encoding krnne string tika int krgnna
data = pd.get_dummies(data)
```

```python
data.head(2)
```

```python
data.info()
```

```python
#Model eka train krnna one wena dewal tika dn krnne
```

```python
x = data.drop('Price_euros', axis=1)
y = data['Price_euros']
```

```python
pip install scikit-learn
```

```python
#testing walatai train krnnai beda gannawa
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25)
```

```python
x_train.shape, x_test.shape
```

```python
#warga kipayakin train krnna hdna nisa mehema function ekak hdgnnwa
def model_acc(model):
    model.fit(x_train,y_train)
    acc = model.score(x_test, y_test)
    print(str(model)+ ' --> '+str(acc))
```

```python
#linear regression model eken train krala balanwa. uda jdapu function eka denwa trin krnnai accuracy eka balagannai
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
model_acc(lr) #me thiynne uda hadapu function eka. meka call krala lr pass krnwa
```

```python
#anith medel wlatath dala balanwa
from sklearn.linear_model import Lasso
lasso = LinearRegression()
model_acc(lasso)

from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
model_acc(dt)

from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
model_acc(rf)
```

```python
#random forest wali train karanna eke parameters balanwa monwda hodama ewa kiyla. eken accuracy eka thwa wadi wenwa
from sklearn.model_selection import GridSearchCV

parameters = {'n_estimators': [10,50,100],
             'criterion':['squard_error','absolute_error','possion']}

grid_obj = GridSearchCV(estimator=rf, param_grid=parameters)

grid_fit = grid_obj.fit(x_train, y_train)

best_model = grid_fit.best_estimator_
best_model
```

```python
#hdagaththa model eka test krnwa
best_model.score(x_test, y_test)
```

```python
#model eka save kragannwa
import pickle
with open('predictor.pickle', 'wb') as file:
    pickle.dump(best_model, file)
```
